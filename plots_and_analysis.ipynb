```python
def plot_l(iterations, value, title='Plot', ylabel='x'):
    plt.figure(figsize=(10, 6))
    plt.plot(iterations, value)  
    plt.yscale('log')             
    plt.ylabel(ylabel)
    plt.xlabel('data points')
    plt.grid(True)
    plt.show()


plot_l(df_h[2], df_h[3], title='hyperparameters variance', ylabel='variance')
plot_l(df_h[2], df_h[4], title='hyperparameters lengthscale', ylabel='lengthscale x1')
plot_l(df_h[2], df_h[5], title='hyperparameters lengthscale', ylabel='lengthscale x2')
plot_l(df_h[2], df_h[6], title='hyperparameters lengthscale', ylabel='lengthscale x3')
plot_l(df_h[2], df_h[7], title='hyperparameters lengthscale', ylabel='lengthscale x4')
```
